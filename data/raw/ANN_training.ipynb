```python
import numpy as np
import pandas as pd
import os
import warnings
warnings.filterwarnings("ignore")
from keras.layers import Dense, Activation
from keras.layers import Dropout
from keras.models import Sequential
from keras.layers import BatchNormalization
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import scipy.stats
from matplotlib.cm import Greys
import matplotlib.style as style
import seaborn as sns
from keras import backend as K
K.set_floatx('float32')
import json
%matplotlib inline
```

```python
N_neurons=20
N_layers=1
```

```python
df=pd.read_csv("fischer_data_processed.csv",sep=",")
```

```python
X = df.iloc[:,0:4].astype('float32')
y = df.iloc[:,4:8].astype('float32')
```

```python
# Initialising the ANN
model = Sequential()

# Adding the input layer and the first hidden layer
if N_layers==1:
     model.add(Dense(N_neurons, activation = 'sigmoid', input_dim = 4))
elif N_layers==2 :
    model.add(Dense(N_neurons, activation = 'sigmoid', input_dim = 4))
    model.add(Dense(units = N_neurons, activation = 'relu'))
elif N_layers==3 :
    model.add(Dense(N_neurons, activation = 'sigmoid', input_dim = 4))
    model.add(Dense(units = N_neurons, activation = 'relu'))
    model.add(Dense(units = N_neurons, activation = 'relu'))
else:  
    print("Check no of layers")

# Adding the output layer
model.add(Dense(units = 4))

model.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics=['mae'])

model.fit(X, y, batch_size = 15, epochs = 800)
```

```python
y_train= model.predict(X)
y, y_train
```

```python
weights = model.get_weights()

# Convert the weights to a list
weights_list = [w.tolist() for w in weights]

# Convert the list to JSON
weights_json = json.dumps(weights_list)

# Save the JSON to a file
with open('fischer_ann_weights.json', 'w') as f:
    f.write(weights_json)
```

```python
# Load the weights from the JSON file
with open('fischer_ann_weights.json', 'r') as f:
    weights_list = json.load(f)

# Convert the weights to NumPy arrays
weights = [np.array(w) for w in weights_list]

# Set the model's weights
model.set_weights(weights)
```
